```python
videos_watched = df['videos_watched'].sort_values(ascending = True)
videos_watched
```

```python
q1 = videos_watched.quantile(0.25)
q3 = videos_watched.quantile(0.75)
q2 = videos_watched.quantile(0.50)
q1,q3,q2
```

```python
iqr = q3-q1
iqr
```

```python
upper_bound = 1.5*iqr + q3
lower_bound = q1- 1.5*iqr

upper_bound, lower_bound
```

```python
videos_clipped =videos_watched.clip(lower = lower_bound, upper = upper_bound)
```

```python
sns.boxplot(videos_clipped)
```
