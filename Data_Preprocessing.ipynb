```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('preprocessing data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['A_median']=df['A'].fillna(df['A'].median())
df
```

```python
b=df['B'].isnull().sum()
b
```

```python

random_sample=df['B'].dropna().sample(b,random_state=1)
random_sample
```

```python
random_sample=df['B'].dropna().sample(b,random_state=13)
random_sample
```

```python
df
```

```python
random_sample.index=df[df['B'].isnull()].index
random_sample
```

```python
df['B']=df['B'].fillna(random_sample)
df
```

```python
# End of distribution method

eod=df['D'].mean()+3*df['D'].std()
eod
```

```python
df['D']=df['D'].fillna(eod)
df
```

```python
# Max Absolute scaling

from sklearn.preprocessing import MaxAbsScaler

mas=MaxAbsScaler()

column=['B']

mas.fit_transform(df[column])

# Fit--> Calcualting parameter of the method

# Transform--> Apply the calucaluted parameter on that column as per formula
```
